```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv(r'E:\Code\PycharmProjects\scikit-learn\resources\Advertising.csv',header=0)
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.describe()
```

```python
Data = data.drop(['Number','sales'],axis=1)
Data.head()
```

```python
sales = data['sales']
type(sales)
sales
```

```python
sales = np.ravel(sales)
type(sales)
sales
```

```python
# 构造训练集和测试集 
X_train,X_test,y_train,y_test=train_test_split(Data,sales,random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 建立线性回归模型
linreg = LinearRegression()
# 将模型拟合到训练数据
linreg.fit(X_train,y_train)

# 打印模型系数
print(linreg.intercept_)
print(linreg.coef_)
```

```python
# 对测试集进行预测
y_pred = linreg.predict(X_test)
print(y_pred)
```

```python
# 模型评估
plt.plot(range(len(y_pred)),y_pred,'b',label='predict')
plt.plot(range(len(y_pred)),y_test,'r',label='test')
plt.legend(loc='upper right')
plt.xlabel('the number of sales')
plt.ylabel('value of sales')
plt.show()
```

```python
# 方差
MSE =metrics.mean_squared_error(y_test,y_pred)
print('方差：',MSE)
```

```python
'''
(1)平均绝对误差(MAE)，对应方法：metrics.mean_absolute_error(true, pred)
(2)均方误差(MSE)，对应方法：metrics.mean_squared_error(true, pred)
(3)均方根误差(RMSE)，对应方法：np.sqrt(metrics.mean_squared_error(true, pred))
'''


# 均方根误差(PMSE)
RMSE = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print('均方根误差：',RMSE)
```
