## 1. QR分解

#### 标准正交矩阵的性质：Q.T * Q = I

#### 如果矩阵各行、各列线性无关，矩阵可逆，则：Q.T=inv(Q)              那么：Q.T是Q的左逆，也是Q的右逆

#### A=QR (Q是标准正交矩阵，R是上三角矩阵)

#### Q.T A=R
#### inv(Q) A=R

demo

    根据 y=ax+b 转化矩阵

    a+b=2

    2a+3b=10
    
    解：a=-4 b=6

  
    
    

```python
A=np.array([[1,1],[2,3]])
b=np.array([[2],[10]])
```

```python
q_a,r_a=la.qr(A)
```

```python
q_a
```

```python
q_a.T 
```

因为q_a是标准正交矩阵 所以 它的转置等于它的逆

```python
la.inv(q_a)
```

Ax=b 

q_a*r_a*x=b

r_a*x=q_a.T*b

```python
r_a
```

```python
q_a.T.dot(b)
```

```python
# 验证
la.solve(A,b)
```
