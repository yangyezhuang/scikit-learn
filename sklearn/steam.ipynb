```python
import pandas as pd
import numpy as np

data = pd.read_csv('../data/zhengqi.txt',sep='\s+')
data.head(5)
```

```python
data.isna()
```

```python
# 查找空值并替换
data['V37'].fillna(data['V37'].mean().round(3),inplace=True)
data.head(5)
```

```python
data.dropna(subset=['target'], inplace=True)
data
```

```python
# 筛选数据
data = data[(data['V0'] < 10) & (data['V0'] > -2)]
data
```
